```python
#Pre-processing data
```

```python
close_price = google_data['Close']
close_price
```

```python
max(close_price.values), min(close_price.values)
```

```python
#Converting this data in the range of 0 to 1 (Makes the training faster)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data= scaler.fit_transform(close_price)
scaled_data
```

```python
x_data = []
y_data = []

for i in range (75, len(scaled_data)):
    x_data.append(scaled_data[(i-75):i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python
splitting_len = int(len(scaled_data) *0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]

x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
x_train.shape
#x is 3D because it represents sequential input data (samples, time steps, features).
```

```python
y_train.shape
#y is 2D because it contains one target value per sample (samples, 1 feature).
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size = 1, epochs = 2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
actual_predictions = scaler.inverse_transform(predictions)
actual_predictions
```

```python
inv_ytest = scaler.inverse_transform(y_test)
inv_ytest
```

```python
rmse = np.sqrt(np.mean((inv_ytest - actual_predictions)**2))
rmse
```

```python
plotting_data = pd.DataFrame(
    {
    'Original Test Data': inv_ytest.reshape(-1),
    'Predictions': actual_predictions.reshape(-1)
    },
    index = google_data.index[splitting_len + 75:]
)
plotting_data.head()
```

```python
plotgraph((15,5), plotting_data, 'Test Data')
```

```python
plotgraph((15,5), pd.concat([close_price[:splitting_len+100], plotting_data], axis=0), 'Whole Data')
```
